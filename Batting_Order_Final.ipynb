```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)

pitching = pd.read_csv('/Users/Brenton/Documents/10day_import/pitch_table.csv', usecols=[0,7,13,14,16,19, 25, 31,57,58])
pitching = pitching.sort_values(by = ['game_id', 'inning'])
```

```python
pitching.head()
```

```python
fp = pitching['pitch_seq']

fp = pd.DataFrame(fp)

fp['first_pitch'] = 0
fp['first_pitch'][fp['pitch_seq'] == 'S'] = 1
fp['first_pitch'][fp['pitch_seq'] == 'C'] = 1
fp['first_pitch'][fp['pitch_seq'] == 'X'] = 1
fp['first_pitch'][fp['pitch_seq'] == 'B'] = 1
fp['first_pitch'][fp['pitch_seq'] == 'F'] = 1
```

```python
fp
```

```python
pitching = pd.merge(pitching, fp, right_index = True, left_index = True)
```

```python
pitching['cum'] = pitching.groupby(['game_id', 'bat_home_id']).first_pitch.cumsum()
pitching['batter_position'] = pitching['cum'] % 9 
pitching['pitch_seq'] = pitching['pitch_seq_x']
pitching = pitching.drop('pitch_seq_x', 1)
pitching = pitching.drop('pitch_seq_y', 1)
```
